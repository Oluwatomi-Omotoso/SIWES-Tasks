Data cleaning using Pandas

The point of this notebook is to clean up the dataset, So that it only contains customers who can be contacted.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
df= pd.read_excel(r'./Customer Call List.xlsx')
df
```

Errors observed within the dataset:

1. Null entries in some columns
2. Inconsistent naming convention.
3. unwanted characters found within entries.

```python
# dropping duplicate rows (if there are any)
df = df.drop_duplicates() 
```

---

Checking column wise for what needs resolving

1. Customer ID Column

```python
# Upon a glance it seems fine, and accurate, but as a good practice, I'll inspect the column.
df['CustomerID'].info()
# It has no missing entries, and all 20 entries are integer datatypes with a consistent naming convention.
```

2. First Name

```python
# Upon a glance, this column also seems fine, but still its worth inspecting.
df['First_Name'].info()
df['First_Name'].describe()
# The column has 20 entries, with 19 of it being unique. All 20 entries have a consistent naming convention
```

3. Last Name

```python
df['Last_Name'].info()
# This column has some issues to it, there's a null entry, and the naming conventions don't tally.
# Removing the unwanted special characters observed in the column
df['Last_Name'] = df['Last_Name'].str.strip("/._")
```

4. Phone Number

```python
df['Phone_Number'].info()
# The phone numbers don't have a standard naming convention, and then there is also some null entries.
```

```python
# clearing out the characters messing up the ideal number convention
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex = True)

# converting the elements in phone number column into strings
df['Phone_Number']= df['Phone_Number'].apply(lambda x: str(x))

# Adding "-" between the numbers for the ideal phone number naming convention
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3]+"-"+x[3:6]+"-"+x[6:10])

# Removing the Na-- and nan-- entries
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')

# Preview
df['Phone_Number']
```

5. Address

```python
# SPlitting the address into three columns: Street address, State and Zip Code.
df['Address'].info()
df[["Street_Address","State","Zip_Code"]] = df["Address"].str.split(",",n=2,expand = True)
```

```python
# dropping the address column
df.drop(columns = "Address", axis = 1, inplace=True)
```

6. Paying Customer

```python
# This column has inconsistent naming conventions.
df['Paying Customer'].info()
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
```

7. Do not Contact

```python
df['Do_Not_Contact'].info()
# Fixing the inconsistent naming conventions
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
```

8. Not Useful Column

```python
# As the name suggests, its not a useful column. So I'll drop it
df = df.drop(columns = 'Not_Useful_Column',axis = 1)
```

```python
df.columns
```

Again, this dataset would have to ignore the customers who don't want to be contacted. 

```python
# Setting the rows with N/a to an empty string
df = df.replace('N/a','')
# Filling the null entries with empty strings.
df = df.fillna('')
```

```python
# Removing customers who don't want to be contacted from the overall dataset
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)
```

```python
# Removing customers without phone numbers
for x in df.index:
    if df.loc[x,"Phone_Number"] == "":
        df.drop(x, inplace = True)
```

```python
df
```

And finally, rewriting the index column, 

```python
df = df.reset_index(drop = True)
```

```python
df
```

```python
# And now, zip code column is completely empty. Might as well drop it.

df.drop(columns= "Zip_Code", axis = 1, inplace=True)
```

Saving the now cleaned up dataset as a csv

```python
df.to_csv('./Cleaned_Customer_Call_List.csv', encoding='utf-8', index = False)
```
